```python
csv_path = '../../data/en.openfoodfacts.org.products.csv'  # Replace with your actual file path

#take this csv file path read 10000 rows sequentially and create a mapping of all additives tags to their names do this until you have processed all the rows in the csv file
chunk_size = 10000
additives_mapping = {}
counter = 0
for chunk in pd.read_csv(csv_path, sep='\t', low_memory=False, chunksize=chunk_size, on_bad_lines="skip"):
    for additives_str in chunk['additives_en'].dropna():
        additives_list = [add.strip() for add in additives_str.split(',')]
        for additive in additives_list:
            additive_array = additive.split('-')
            if len(additive_array) == 2:
                tag, name = additive_array
                tag = tag.lower().strip()
                name = name.strip()
                if tag not in additives_mapping:
                    additives_mapping[tag] = name

    counter += len(chunk)
    print(f"Processed {counter} rows...")
```

```python
with open('../../data/processed/additives_mapping.json', 'w') as f:
    json.dump(additives_mapping, f, indent=4)
```
